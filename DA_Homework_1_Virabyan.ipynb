```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
import scipy.stats as stats
import math
```

Придумаем функцию для некоего распределения

```python
def f(x):
    if -3 < x < -2:
        return 6*x+20
    elif -2 < x < 1:
        return x+10
    elif 1 < x < 2:
        return x/2+20
    elif 2 < x < 3:
        return -4*x+20
    else:
        return 0
```

Сгенерируем массив чисел из этой функции, построим график плотности вероятности и гистограмму

```python
x = pd.Series(np.arange(-4, 4, 0.01))
y = x.apply(lambda x: f(x))
plt.plot(x, y)
plt.show()
```

```python
X = stats.uniform(loc=-3, scale = 6).rvs(size=100000)
Y = stats.uniform(loc=0, scale = 21).rvs(size=100000)
```

```python
Z = []
for i in range(len(X)):
    if Y[i] < f(X[i]):
        Z.append(X[i]) 
```

```python
ax = sns.displot(Z,
                  bins=10,
                  kde=True,
                  color='red'
                  )
ax.set()
```

```python
distribution_discrete = stats.binom(n = 5, p = 0.85)
```

```python
cdf_bi = distribution_discrete.cdf
x = np.arange(-1, 6, 0.01)
plt.plot(x, cdf_bi(x))
plt.show()
```
